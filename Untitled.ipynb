```python
from collections import Counter

import numpy as np
import pandas as pd
import seaborn as sns
from tqdm.auto import tqdm
tqdm.pandas()
```

```python
df = pd.read_csv("train.sep.txt", sep="\t", header=None)
df.columns = ["file", "pinyin", "text"]
df
```

```python
vocabulary = Counter(df.pinyin.str.cat().strip().split(" "))
vocabulary.most_common(30)
```

```python
idx2syl = list(vocabulary.keys())
syl2idx = {syl: idx for idx, syl in enumerate(idx2syl)}
```

```python
bigrams = np.zeros((len(idx2syl), len(idx2syl)))
for _, row in tqdm(df.iterrows(), total=len(df)):
    toks = row.pinyin.strip().split(" ")
    for tok1, tok2 in zip(toks, toks[1:]):
        bigrams[syl2idx[tok1], syl2idx[tok2]] += 1
```

```python
bigram_df = pd.DataFrame(bigrams)
bigram_df.index = idx2syl
bigram_df.index.name = "tok1"
bigram_df.columns = idx2syl
bigram_df.columns.name = "tok2"
bigram_df = pd.DataFrame(bigram_df.unstack().reset_index(name="count"))
bigram_df
```

```python
bigram_df[["tok1_base", "tok1_tone"]] = bigram_df.tok1.str.extract("([^0-9]+)([0-9])")
bigram_df[["tok2_base", "tok2_tone"]] = bigram_df.tok2.str.extract("([^0-9]+)([0-9])")
```

```python
bigram_df["tok1_tone"] = bigram_df.tok1_tone.astype(int)
bigram_df["tok2_tone"] = bigram_df.tok2_tone.astype(int)
```

```python
reverse_2vs3_df = bigram_df[bigram_df.tok1_tone.isin((2, 3))] \
    .pivot_table(columns="tok1_base", index=["tok1_tone", "tok2"], values="count").fillna(0)
total_counts = reverse_2vs3_df.groupby("tok2").sum().unstack()
reverse_2vs3_df = (reverse_2vs3_df.loc[2] / reverse_2vs3_df.groupby("tok2").sum()).fillna(0)
reverse_2vs3_df = pd.DataFrame({"p2": reverse_2vs3_df.unstack(), "total": total_counts})
reverse_2vs3_df
```

```python
sns.displot(reverse_2vs3_df[reverse_2vs3_df.p2.between(0.1, 0.9)].p2)
```
